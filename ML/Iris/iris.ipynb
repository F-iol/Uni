```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = load_iris()
```

```python
iris_df = pd.DataFrame(data.data,columns=data.feature_names)
```

```python
iris_df['target'] = data.target
iris_df['target_name'] = iris_df['target'].apply(lambda x :data.target_names[x])
```

```python
iris_df
```

```python
from sklearn.model_selection import KFold,train_test_split
```

```python
kfold = KFold(n_splits=5)
```

```python
x = iris_df.drop(columns=['target','target_name'])
y = iris_df['target']
```

```python
x_tr ,x_test,y_tr,y_test=train_test_split(x,y,test_size=.2,random_state=69)
```

```python
for train_idxs , val_idxs in kfold.split(x_tr):
    x_train ,x_val =x_tr.iloc[train_idxs] , x_tr.iloc[val_idxs]
    y_train,y_val = y_tr.iloc[train_idxs] , y_tr.iloc[val_idxs]
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(random_state=69,n_estimators=100,n_jobs=-1,max_features=3,min_samples_split=18)
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
def try_params(**params):
    model=RandomForestClassifier(random_state=69,n_jobs=-1,**params)
    model.fit(x_train,y_train)
    return print(f'Training acc: {accuracy_score(y_train,model.predict(x_train))}   Val acc: {accuracy_score(y_val,model.predict(x_val))}')
```

```python
print(f'Training acc: {accuracy_score(y_train,model.predict(x_train))}   Val acc: {accuracy_score(y_val,model.predict(x_val))}')
```

```python
try_params(n_estimators=100,min_samples_split=18,min_samples_leaf=1,max_features='sqrt')
```

```python
results = []

for param_value in range(1,20):
    model=RandomForestClassifier(random_state=69,n_estimators=100,n_jobs=-1,max_features=3,min_samples_split=18)
    model.fit(x_train,y_train)
    score = accuracy_score(y_val,model.predict(x_val))
    results.append({'value':param_value,'val_acc':score})
    
result_df=pd.DataFrame(results)
result_df
```
